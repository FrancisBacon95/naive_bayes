```python
contents_thon=df_thon['content']
contents_thon=contents_thon.apply(lambda x: re.sub(r'[^ ㄱ-ㅣ가-힣]+',"",x))
contents_thon
```

```python
contents=df_2019['content']
contents=contents.apply(lambda x: re.sub(r'[^ ㄱ-ㅣ가-힣]+',"",x))
contents
```

```python
titles_thon=df_thon['title']
titles_thon=titles_thon.apply(lambda x: re.sub(r'[^ ㄱ-ㅣ가-힣]+',"",x))
```

```python
titles=df_2019['title']
titles=titles.apply(lambda x: re.sub(r'[^ ㄱ-ㅣ가-힣]+',"",x))
```

```python
okt=Okt()
word_dic={}
for lines in titles:
    malist=okt.pos(lines)
    for word in malist:
        if word[1] == "Noun":
            if not(word[0] in word_dic):
                word_dic[word[0]] = 0
            word_dic[word[0]] += 1
keys = sorted(word_dic.items(), key=lambda x:x[1],reverse=True) 
for word, count in keys[:100]:
    print("{0}({1})".format(word,count),end="\n")
print()
```
